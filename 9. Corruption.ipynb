```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel("9. Corruption.xls")
df.shape
```

```python
df.head()
```

```python
# Scatter plot
sns.scatterplot(data=df,
                x = "gdp_cap",
                y = "corruption index*",
                size = "corruption index*"
               )
```

```python
#regression plot
sns.regplot(data=df,
            x = "gdp_cap",
            y = "corruption index*"
               )
```

```python
#lm plot
sns.lmplot(data=df,
           x = "gdp_cap",
           y = "corruption index*"
          )
```
