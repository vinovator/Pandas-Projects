```python
data = pd.read_excel("1. Medical-Cancer.xls", sheet_name="Sheet1")
data.shape
```

```python
#seaborn module
sns.scatterplot(data=data, x="numer of cigar.", y="bladder cancer")
# Cant find a way to introduce seconday y axis
```

```python
#Unpivot type of cancer into the rows inorder to create scatter plots with hues
data1 = data.melt(id_vars=["numer of cigar."], 
                  var_name = "Type of cancer",
                  value_name = "Cancer value"
                 )
data1
```

```python
#Seaborn plotting using hue and style to differentiate between the types of cancers
sns.scatterplot(data=data1, 
                x="numer of cigar.", 
                y="Cancer value", 
                hue="Type of cancer", 
                style="Type of cancer",
                size="Cancer value"
               )
```
